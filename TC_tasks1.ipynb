```python
#dropping lyrics column as it contains null values
#dropping unnamed column and uri column
```

```python
df.drop(['Unnamed: 0','lyrics','URI'],axis=1,inplace=True)
```

```python
#dropping duplicate rows
```

```python
temp=df[df['Title'].duplicated()]
```

```python
#finding unique songs
```

```python
list_of_songs=list(temp.Title.unique())
```

```python
# filtering songs with same title but by different artists
```

```python
n=[]
for i in range (len(list_of_songs)):
    m=(list(df[df['Title']==list_of_songs[i]]['Artist'])) # getting all artists of duplicate songs
    if m[0]==m[1]:
        pass
    else:
        j=(df[df['Title']==list_of_songs[i]].index) # if the artist is different , get the index of first artist
        n.append(j[0])
```

```python
df1=df.iloc[n]
```

```python
#dropping duplicate values
df2=df.drop_duplicates(subset='Title',keep='last')
```

```python
#concatenating two dataframes

data=pd.concat([df1,df2])
```

```python
data.reset_index(inplace=True,drop=True)
```
